```python
# remap for plot labels to show up
train['CS'] = train['Computer Science'].map({0:'not CS', 1:'CS'})
train['PH'] = train['Physics'].map({0:'not PH', 1:'PH'})
train['MA'] = train['Mathematics'].map({0:'not MA', 1:'MA'})
train['ST'] = train['Statistics'].map({0:'not ST', 1:'ST'})
train['BIO'] = train['Quantitative Biology'].map({0:'not BIO', 1:'BIO'})
train['FI'] = train['Quantitative Finance'].map({0:'not FI', 1:'FI'})
```

# Scatterplot

```python
# compute PCA components in batch to prevent system crash, default n=2
pca = pd.Series(dtype=float)
batch = 5000

for i in range(len(train)//batch+1):
    if i < len(train)//batch:
        pca_batch= train['processed'][i*batch:(i+1)*batch].pipe(hero.tfidf).pipe(hero.pca)
    else:
        pca_batch= train['processed'][i*batch:].pipe(hero.tfidf).pipe(hero.pca)
    pca = pd.concat([pca, pca_batch])

train['pca'] = pca.values
```

```python
# plot PCA components
cols = ['CS','PH','MA','ST','BIO','FI'] 
labels = ['Computer Science','Physics','Mathematics','Statistics','Quantitative Biology','Quantitative Finance']

for col, label in zip(cols, labels):
    hero.scatterplot(train, 'pca', color=col, title=label)
```

Observation: Quantitative Biology and Quantitative Finance appear very scattered on the PCA plots due to their low counts. As with any classification task, predictions on minority classes are expected to be poor.
